```python
import numpy as np

import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(20, 10))

multiplier = 2
a = np.random.randn(100000)
b = a * multiplier

if multiplier > 1:
    plt.hist(b, bins=1000, label='multiplied')
    plt.hist(a, bins=1000, label='original')
else:
    plt.hist(a, bins=1000, label='original')
    plt.hist(b, bins=1000, label='multiplied')

c = np.random.rand(100000)
plt.hist(c, bins=1000, label='uniform')

plt.title('Distributions visualization')
plt.xlabel('Values')
plt.ylabel('Counts')
plt.legend();
```

```python
plt.figure(figsize=(20, 10))

multiplier = 2
a = np.random.dirichlet([0.2, 0.3], size=100000)

print(a)
plt.hist(a, bins=1000)

plt.title('Distributions visualization')
plt.xlabel('Values')
plt.ylabel('Counts');
```
