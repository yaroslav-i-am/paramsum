```python
import pandas as pd

from razdel import tokenize, sentenize
from pymorphy2 import MorphAnalyzer
from nltk.corpus import stopwords

from tqdm.notebook import tqdm

from pandarallel import pandarallel
```

```python
reviews = pd.read_csv('./abae_pytorch/data/reviews_Review_Label.csv')['review']
reviews
```

```python
reviews.to_csv('./abae_pytorch/data/test_csv.csv', index=False)
```

```python
pandarallel.initialize(nb_workers=12, progress_bar=True)

morph = MorphAnalyzer(lang='ru')
rus_stopwords = stopwords.words("russian")
print(rus_stopwords[:10])

def parse(review, morph, tokenize, sentenize, rus_stopwords):    
    sentences = [sent.text for sent in list(sentenize(review))]
    tokens = [[morph.parse(token.text)[0].normal_form for token in list(tokenize(sent)) if token.text not in rus_stopwords and str.isalpha(token.text)] for sent in sentences]

    return tokens

# for review in tqdm(reviews):
#     sentences = [sent.text for sent in list(sentenize(review))]
#     # print(sentences)
#     tokens = [[morph.parse(token.text)[0].normal_form for token in list(tokenize(sent)) if token.text not in rus_stopwords and str.isalpha(token.text)] for sent in sentences]

#     # print(tokens)

#     # break
```

```python
tok_reviews = reviews.parallel_apply(parse, args=(morph, tokenize, sentenize, rus_stopwords))
```

```python
tok_reviews
```

```python
with open('./abae_pytorch/data/kp_reviews.txt', 'wt', encoding='UTF-8') as f:
    for review in tok_reviews:
        for sent in review:
            f.write(' '.join(sent) + '\n')
```
