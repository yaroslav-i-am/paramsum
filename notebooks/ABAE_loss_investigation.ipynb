```python
import pandas as pd

import re

import matplotlib.pyplot as plt
import seaborn as sns

from IPython.display import display
```

```python
def plot_losses(losses: pd.DataFrame) -> None:
    plt.figure(figsize=(10, 10))
    
    for i in losses.columns:
        plt.plot(losses[i], label=f'{i} aspects')
    
    plt.title('Train losses after each epoch')
    plt.xlabel('Epoch number')
    plt.ylabel('Loss ratio')
    plt.legend()
    plt.show()
```

```python
sim_losses = pd.read_excel('../data/losses/sim_loss.xlsx')
print((sim_losses.dtypes == 'float64').all())
display(sim_losses)

plot_losses(sim_losses)
```

```python
ort_losses = pd.read_excel('../data/losses/ort_loss.xlsx')
print((ort_losses.dtypes == 'float64').all())
display(ort_losses)

plot_losses(ort_losses)
```
