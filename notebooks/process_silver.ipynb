```python
import pandas as pd
import numpy as np

import re

from tqdm.notebook import tqdm, trange
from IPython.display import display
```

```python
gold = pd.read_csv('../data/working_dir/ready_markup.csv')
gpt_markup = pd.read_csv('../data/working_dir/gpt_markup.csv')

mix = gold.copy()
mix['gpt_markup'] = gpt_markup

display(gold)
display(gpt_markup)
display(mix)
```

```python
def process_gpt_markup(line: str):
    print('INITIAL LINE')
    print(line)

    if line in (None, np.nan) or 'None' in line:
        line = np.nan
    else:
        line = list(filter(lambda el: el if len(el) > 0 else None, line.split('\n')))
        line = list(map(lambda el: re.sub(r'^\* ', r'', el), line))
        line = list(map(lambda el: re.sub(r'^\- ', r'', el), line))
        line = list(map(lambda el: re.sub(r'^\— ', r'', el), line))
        line = list(map(lambda el: el.strip(), line))


        line = set(line)

    
    print('FINAL LINE')
    print(line)
    print('\n-----------------------------------------------\n')
    
    return line

mix['gpt_markup_preprocessed'] = mix['gpt_markup'].apply(process_gpt_markup);
```
