```python
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns

from tqdm.notebook import trange, tqdm

from ast import literal_eval
tqdm.pandas()
```

```python
gold388 = pd.read_csv('../data/working_dir/crowd_markup.csv')
gold99 = pd.read_csv('../data/working_dir/ready_markup.csv')

gold = pd.concat([gold99, gold388], axis=0)

gold
```

```python
pd.read_csv('../data/working_dir/saiga/markups/markup_T-None.csv')
```

```python
print(gold.aspect.value_counts())
g = sns.barplot(y=gold.aspect.value_counts(), x=gold.aspect.value_counts().index, palette='Blues')
g.set_xticklabels(labels=gold.aspect.value_counts().index, rotation=20, horizontalalignment='right')
plt.title('Aspects distribution in whole Labelled Dataset')
plt.ylabel('Count')

plt.savefig('../data/barplot_full_gold.aspect.value_counts.png', bbox_inches='tight', dpi=400)
```
